```
import pandas as pd
import numpy as np
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('iris.csv')
```

```
df.shape
```

```
df['species'].value_counts()
```

```
df.plot(kind="scatter", x='petal_length', y="petal_width")
plt.grid()
```

```
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="species",size=12)\
.map(plt.scatter,"petal_length","petal_width")\
.add_legend()
```

```
plt.close()
```

```
sns.set_style("whitegrid")
sns.pairplot(df,hue="species",size=5)
```

```
iris_setosa = df.loc[df["species"]=="setosa"]
iris_virginica = df.loc[df["species"]=="virginica"]
iris_versicolor = df.loc[df["species"]=="versicolor"]
```

```
plt.plot(iris_setosa["petal_length"],np.zeros_like(iris_setosa["petal_width"]))
plt.plot(iris_virginica["petal_length"],np.zeros_like(iris_virginica["petal_width"]))
```

```
plt.close()

sns.FacetGrid(df,hue="species", size=5) \
.map(sns.distplot,"petal_length") \
.add_legend()
```

```
density, edges =np.histogram(iris_setosa["petal_length"],bins=10,density=True)
```

```
pdf = density/np.sum(density)
cdf = np.cumsum(pdf)
```

```
densityv, edgesv =np.histogram(iris_versicolor["petal_length"],bins=10,density=True)
```

```
pdfv = densityv/np.sum(densityv)
cdfv = np.cumsum(pdfv)
```

```
densityv1, edgesv1 =np.histogram(iris_virginica["petal_length"],bins=10,density=True)
pdfv1 = densityv1/np.sum(densityv1)
cdfv1 = np.cumsum(pdfv1)
```

```
plt.plot(edges[1:],cdf)
plt.plot(edges[1:],pdf)

plt.plot(edgesv[1:],cdfv)
plt.plot(edgesv[1:],pdfv)

plt.plot(edgesv1[1:],cdfv1)
plt.plot(edgesv1[1:],pdfv1)
```

```
surv1 = df.loc[df["Survival_status"]==1]
```

```
surv2 = df.loc[df["Survival_status"]==2]
```

```
c1,e1 = np.histogram(surv1,density=True)
pdf1 = c1/np.sum(c1)
cdf1= np.cumsum(pdf1)
```

```
c2,e2 = np.histogram(surv2,density=True)
pdf2 = c2/np.sum(c2)
cdf2= np.cumsum(pdf2)
```
